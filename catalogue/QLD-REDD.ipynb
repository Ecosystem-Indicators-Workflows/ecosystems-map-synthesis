# Queensland Regional Ecosystems

> *During the 1990s the regional ecosystems (RE) classification framework was developed and widely adopted in Queensland to assist in planning, regulation and management for biodiversity, both on and off conservation reserve estate... Updated regional ecosystem descriptions in the format of Sattler and Williams (1999) are maintained in REDD.* [@Nelder_2023]

The Regional Ecosystem Description Database or REDD is available online [@REDD_2024]. It contains descriptions of more than 1,200 REs in Queensland.

Assignment of RE to the ecosystem functional groups (level 3 of the GET) is currently based on preliminary analysis by experts following cross-referencing guidelines [@IUCN_2025].

## Download spatial data

The *Biodiversity status of 2021 remnant regional ecosystems – Queensland* dataset was published by the Department of Environment, Science and Innovation. This is a foundational dataset that provides a geographical representation of the 2021 Remnant Regional Ecosystems at a map scale of 1:100,000 and 1:50,000, 1:25,000 in part, based on surveys of vegetation communities and related landform, soils and geology.

We downloaded Version 13.1 from 2021 from the [Queensland Spatial Catalogue](https://qldspatial.information.qld.gov.au/catalogue/custom/detail.page?fid={8FDF54D2-654C-4822-8295-1D8E8E772373}), and copied the spatial data into a local folder for further processing.

::: {.callout-caution}
Data download from Queensland Spatial Catalogue requires filling a form to request a link for downloading. This link is sent via email and is only accessible for a limited time. This data source cannot be integrated into this reproducible workflow. 
:::

````bash
ls $REPODIR/gisdata/QSC_Extracted_Data_20250827_210509181718-10492/
````

````bash
ogrinfo -al -so $REPODIR/gisdata/QSC_Extracted_Data_20250827_210509181718-10492/data.gpkg |head -n 20
````
