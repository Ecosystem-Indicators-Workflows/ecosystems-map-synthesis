# Download indicative maps

In this document we will use the spatial data of the [***Ecological Communities of National Environmental Significance Distributions***](https://fed.dcceew.gov.au/datasets/3ce27fab85194f468bcfb106e5be15ee/about) dataset provided by the Department of Climate Change, Energy, the Environment and Water in the Find Environmental Data geospatial catalogue.

The instructions below provide all the steps to download the data and metadata from the catalogue and prepare it for further analysis. 

The input data is in a compressed Geodatabase (zipped GDB) format and is provided with a Creative Commons License ([CC BY 4.0](https://creativecommons.org/licenses/by/4.0)).

We filter the data to extract the ecological community of interest and export this in Geopackage (GPKG) format.

## Relative path

```bash
REPODIR=$(pwd)
```

```bash
export OUTPUT_DIR=$REPODIR/../gisdata/TEC
mkdir -p $OUTPUT_DIR
```

## Download from data repository

```bash
wget --continue https://www.arcgis.com/sharing/rest/content/items/3ce27fab85194f468bcfb106e5be15ee/data --output-document=$OUTPUT_DIR/ECNes_public_gdb.zip
```

Unzip data

```bash
unzip -u $OUTPUT_DIR/ECNes_public_gdb.zip -d $OUTPUT_DIR
```

## Metadata

The downloaded zipfile does not include metadata, but this is available from the [FED website](https://www.arcgis.com/sharing/rest/content/items/3ce27fab85194f468bcfb106e5be15ee/info/metadata/metadata.xml?format=default).

```bash
URL="https://www.arcgis.com/sharing/rest/content/items/3ce27fab85194f468bcfb106e5be15ee/info/metadata/metadata.xml?format=default"

wget --continue $URL --output-document=$OUTPUT_DIR/metadata.xml
```

 After manual inspection I found that this file contains fields for purpose, abstract and credit. We can automatically query the downloaded file using `xmllint` and `xpath` expressions to extract this information. ^[Here is useful to add the `local-name()` to ignore the namespaces that would otherwise cause a "XPath set is empty" error.]

```bash
XPATH="////*[local-name()='purpose']/*[local-name()='CharacterString']/text()"
xmllint $OUTPUT_DIR/metadata.xml  --xpath $XPATH | fmt -w 80
```

```bash
XPATH="////*[local-name()='abstract']/*[local-name()='CharacterString']/text()"
xmllint $OUTPUT_DIR/metadata.xml  --xpath $XPATH | fmt -w 80
```

```bash
XPATH="////*[local-name()='credit']/*[local-name()='CharacterString']/text()"
xmllint $OUTPUT_DIR/metadata.xml  --xpath $XPATH | fmt -w 80
```

## Extract ecological community of interest

Now we want to extract the data for the target community. Here we use functions from the `gdal` library. First we use `ogrinfo` to get a list of all ecological communities in the dataset:

```bash
ogrinfo $OUTPUT_DIR/ECNes_public_gdb.gdb -sql "SELECT distinct COMMUNITY FROM ECnes_public where COMMUNITY like '%Fleurieu%'"
```

And now we extract it into a new vector file using `ogr2ogr`: 

```bash
ogr2ogr $OUTPUT_DIR/Swamps-Fleurieu-IM.gpkg $OUTPUT_DIR/ECNes_public_gdb.gdb ECnes_public -where "COMMUNITY='Swamps of the Fleurieu Peninsula'" -nlt "MULTIPOLYGON" -makevalid
```
